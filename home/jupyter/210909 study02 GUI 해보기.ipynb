```python
##
#	온도 변환 프로그램을 GUI 기반으로 구현한다. 
#
from tkinter import *
import tkinter.font

# 이벤트 처리 함수를 정의한다. 
def process():
    tf = float(e1.get())		# e1에서 문자열을 읽어서 부동소수점형으로 변경
    tc = (tf-32.0)*5.0/9.0		# 화씨 온도를 섭씨 온도로 변환한다. 
    e2.delete(0, END)			# 처음부터 끝까지 지운다.
    e2.insert(0, round(tc,1))		# tc 변수의 값을 문자열로 변환하여 추가한다.
    
    
window  = Tk()
window.title("김민상")
window.geometry("250x150+100+100")
font=tkinter.font.Font(family="맑은 고딕", size=20)

Label(window , text="화씨", font=font).grid(row=0, column=0)
Label(window, text="섭씨", font=font).grid(row=1, column=0)

e1 = Entry(window)
e2 = Entry(window)
e1.grid(row=0, column=1)
e2.grid(row=1, column=1)

e1.place(x = 70,y = 10,width=100, height=30)
e2.place(x = 70,y = 50,width=100, height=30)

Button(window, text="화씨->섭씨", command=process).grid(row=2, column=1)
window.mainloop()
```

```python
from tkinter import *
import random

answer = random.randint(1,100)		# 정답을 1에서 100 사이의 난수로 설정한다. 

def guessing():
	guess = int(guessField.get())	# 텍스트 필드에서 사용자가 입력한 값을 가져온다.

	if   guess > answer:
		msg = "높음!"
	elif   guess < answer:
		msg = "낮음!"
	else:
		msg = "정답!"
		
	resultLabel["text"] = msg		# 메시지를 출력한다. 
	guessField.delete(0, 5)

def reset():				# 정답을 다시 설정한다. 
	global answer
	answer = random.randint(1,100)
	resultLabel["text"] = "다시 한번 하세요!"

window = Tk()
window.configure(bg="white")
window.title("숫자를 맞춰보세요!")
window.geometry("430x90")

titleLabel = Label(window, text="숫자 게임에 오신 것을 환영합니다!", bg="white")
titleLabel.pack()

guessField = Entry(window)
guessField.pack(side="left")
tryButton = Button(window, text="시도", fg="green", bg="white", command=guessing )
tryButton.pack(side="left")

resetButton = Button(window, text="초기화", fg="red", bg="white", command=reset)
resetButton.pack(side="left")

resultLabel = Label(window, text="1부터 100사이의 숫자를 입력하시오.", bg="white") 
resultLabel.pack(side="left")

window.mainloop()
```
