```python
ebola = pd.read_csv('../data/country_timeseries.csv')
print(ebola.count())
```

```python
ebola.columns
```

```python
# 전체 데이타 확인
ebola.shape
```

```python
#첫열의 갯수 확인
ebola.shape[0]
```

```python
# 카운트가 몇개인지 알아봄
ebola.count()
```

```python
# 누락값을 가져옴
num_rows = ebola.shape[0]
num_missing = num_rows - ebola.count() 
print(num_missing)
```

```python
# 누락값 확인하는것
import numpy as np 
# 넘파이를 활용해야함
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
```

```python
# 전체 갯수 세는것

print(np.count_nonzero(ebola.isnull()))
```

```python
# 세어보기 
print(ebola.Cases_Senegal.value_counts(dropna=False))
```

```python
# 자료가 없을때NAN 0을 채우는 방법
print(ebola.fillna(0).iloc[0:10, 0:5])
```

```python
# 자료가 없을때는 바로 전값을 가져오는 방법
print(ebola.fillna(method='ffill').iloc[0:10, 0:5])
```

```python
# 자료가 없을때는 바로  뒷값을 가져오는 방법
print(ebola.fillna(method='bfill').iloc[0:10, 0:5])
```

```python
# 위 아래값을 평균을 내서 해주는값
print(ebola.interpolate().iloc[0:10, 0:5])
```
